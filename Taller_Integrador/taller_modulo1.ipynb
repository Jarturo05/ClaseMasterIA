### 1.1 — Listas y operaciones básicas

Tienes la siguiente lista con la esperanza de vida de 10 países (en años).

````python
# Lista dada — NO modificar
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
````

**Ejercicio 1.1** — Usando Python puro (sin Pandas ni funciones built-in como `sum()` o `max()`):
- Calcula el **promedio** de la lista usando un bucle `for`
- Encuentra el **valor máximo** usando un bucle `for`
- Encuentra el **valor mínimo** usando un bucle `for`

````python
# Tu código aquí

# Promedio
suma = 0

for n in esperanzas_de_vida:
    suma += n
promedio = suma /len(esperanzas_de_vida)
print(f"El promedio de las esperazanas de vida es: {promedio:.2f}")
 
# Máximo
maximo = esperanzas_de_vida[0]

for n in esperanzas_de_vida:
    if n > maximo:
        maximo = n
print(f"El número máximo de la lista es: {maximo}")

# Mínimo

minimo = esperanzas_de_vida[0]

for n in esperanzas_de_vida:
    if n < minimo:
        minimo = n
print(f"El número minimo es: {minimo}")
````

### 1.3 — Funciones

Las funciones nos permiten reutilizar lógica.

**Ejercicio 1.3** — Crea una función llamada `clasificar_esperanza_vida` que reciba un número (años) y retorne:
- `"Alta"` si es >= 75
- `"Media"` si es entre 60 y 74.9
- `"Baja"` si es < 60

Prueba la función con al menos 5 valores distintos e imprime los resultados.

````python
# Tu código aquí
def clasificar_esperanza_vida(años):     
    if años >= 75:           
        return "Alta"   
    elif años > 60 and años < 74.9:
        return "Media"        
    elif años < 60 and años > 0:
        return "Baja"        
    else:
        años <= 0
        return "El valor no puede ser negativo o igual a 0"
                
        
edad = 5
print(f"{clasificar_esperanza_vida(edad)}")

edad = 73
print(f"{clasificar_esperanza_vida(edad)}")

edad = 15
print(f"{clasificar_esperanza_vida(edad)}")

edad = -5
print(f"{clasificar_esperanza_vida(edad)}")

edad = 50
print(f"{clasificar_esperanza_vida(edad)}")
````

### 2.3 — Estadísticas descriptivas

🔧 **Ejercicio 2.3**
1. Muestra el resumen estadístico completo del dataset
2. ¿Cuál es la esperanza de vida **promedio** mundial?
3. ¿Cuál es el país con **mayor** gasto en salud? ¿Y el de **menor**?
4. ¿Cuál es la **mediana** de la mortalidad infantil?

````python
# Tu código aquí

#Esperanza de vida promedio mundial
promedio = df["esperanza_vida"].mean()
print(f"El Promedio mundial basado en la esperanza de vida es de: {promedio}")

#Mayor gasto en salud
mayor = df["gasto_salud_usd"].idxmax()
datos_mayor = df.loc[mayor, ["pais", "gasto_salud_usd"]]
print(f"\nEl Pais con mayor gasto en salud es: ")
print(datos_mayor)

#Menor gasto en salud
menor = df["gasto_salud_usd"].idxmin()
datos_menor = df.loc[menor, ["pais", "gasto_salud_usd"]]
print(f"\nEl Pais con menor gasto en salud es: ")
print(datos_menor)

#La mediana de la mortalidad infantil
mediana = df["mortalidad_infantil"].median()
print(f"\nLa medianda de la mortalidad infantil es: {mediana}")
````

### 2.4 — Filtrado y selección

🔧 **Ejercicio 2.4**
1. Filtra los países de **América** (`region == "Americas"`) y muestra solo las columnas: `pais`, `esperanza_vida`, `gasto_salud_usd`
2. Filtra los países con esperanza de vida **mayor a 80 años** — ¿cuántos son? ¿De qué regiones son?
3. Filtra los países con **nivel_ingresos == "Low"** y calcula su promedio de `mortalidad_infantil`
4. ¿Cuántos países tienen vacunación por encima del 90%?

````python
# Tu código aquí

#Filtrar los paises de America
paises_america = df.loc[df["region"] == "Americas", ["pais", "esperanza_vida", "gasto_salud_usd"]]
print("Los paises de America son: ")
print(paises_america)

#Filtrar los paises con esperanza de vida mayor a 80 años
paises_esperanza_vida_80 = df.loc[df["esperanza_vida"] > 80, ["region"]]
cantidad = paises_esperanza_vida_80.shape[0]
conteo_region = paises_esperanza_vida_80.value_counts().reset_index()
conteo_region.columns = ["region", "cantidad"]

print("\nEl conteo de la region basado en los paises con esperanza de vida mayor a 80 es de:")
print(conteo_region)
print(f"\nCantidad de paises: {cantidad}")

#Filtrar los paises con nivel de ingresos "Low" y calcular su promedio de mortalidad infantil
nivel_ingresos_low = df.loc[df["nivel_ingresos"] == "Low", "mortalidad_infantil"]
promedio_low = nivel_ingresos_low.mean()

print(f"\nLos paises con nivel de ingreso ´low´ tiene un promedio de mortalidad infantil de: {promedio_low:.5f}")

#Vacunacion por encima del 90%
vacunacion_90 = df.loc[df["vacunacion_pct"] > 90]
cantidad_vacunacion = vacunacion_90.shape[0]

print(f"\nEl numero de paises que tienen la vacunacion por encima del 90% es: {cantidad_vacunacion}")
````

### 2.5 — Agrupaciones

🔧 **Ejercicio 2.5** — Usa `groupby()` para:
1. Calcular la **esperanza de vida promedio** por `region`
2. Calcular el **gasto en salud promedio** por `nivel_ingresos`
3. Calcular el **número de países** por región
4. ¿Qué región tiene la mayor esperanza de vida promedio? ¿Y la menor?

````python
# Tu código aquí

#Esperanza de vida promedio por region
esperanza_vida_region = df.groupby("region")["esperanza_vida"].mean()
print("La esperanza de vida promedio por region es: ")
print(esperanza_vida_region)

#Gasto en salud promedio por nivel de ingresos
gasto_salud_promedio_ingresos = df.groupby("nivel_ingresos")["gasto_salud_usd"].mean()
print("\nEl gasto en salud promedio basado en los ingresos es: ")
print(gasto_salud_promedio_ingresos)

#Numero de paises por region
numero_paises_region = df.groupby("region")["pais"].count()
print("\nEl numero de paises por region: ")
print(numero_paises_region)

#Valor promedio de region con mayor esperanza de vida
region_promedio_esperanza_vida = df.groupby("region")["esperanza_vida"].mean()
region_mayor = region_promedio_esperanza_vida.idxmax()
valor_mayor = region_promedio_esperanza_vida.max()

print(f"\nLa region que tiene mayor esperanza de vida es: {region_mayor}")
print(f"Con un valor de: {valor_mayor:.5f}")

#Valor promedio de region con menor esperanza de vida
region_promedio_esperanza_vida = df.groupby("region")["esperanza_vida"].mean()
region_menor = region_promedio_esperanza_vida.idxmin()
valor_menor = region_promedio_esperanza_vida.min()

print(f"\nLa region que tiene menor esperanza de vida es: {region_menor}")
print(f"Con un valor de: {valor_menor:.5f}")
````

### 2.6 — Nueva columna calculada

🔧 **Ejercicio 2.6**
1. Crea una nueva columna llamada `categoria_vida` aplicando la función `clasificar_esperanza_vida` que creaste en la Parte 1 a la columna `esperanza_vida`
2. ¿Cuántos países caen en cada categoría? (usa `value_counts()`)
3. ¿Qué porcentaje del total representa cada categoría?

````python
# Tu código aquí
#Se crea una nueva columna donde se agrega la funcion antes realizada y se muestra como queda con respecto a la esperanza de vida
df["categoria_vida"] = df["esperanza_vida"].apply(clasificar_esperanza_vida)
print(df[["pais", "esperanza_vida", "categoria_vida"]])

#Numero de paises que caen en cada categoria
paises_categoria_vida = df["categoria_vida"].value_counts()
print(f"\nEl numero de paises en cada categoria son: ")
print(paises_categoria_vida)

#Porcentaje total que representa cada categoria
porcentaje_categoria_vida =paises_categoria_vida / paises_categoria_vida.sum() *100

print(f"\nEl porcentaje calculado para cada pais es: ")
print(porcentaje_categoria_vida)
````
